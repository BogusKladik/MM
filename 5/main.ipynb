```python
import matplotlib.pyplot as plt
import numpy as np
import condition
from methods.dichotomy import dichotomy_method
from methods.chord import chord_method, chord_method1
from methods.simple_iterations import simple_iterations_method
from methods.newton import newton_method
```

```python
print(f"Исходная функция: {condition.function_as_string()}")
a = float(input("Введите переменную a: "))  # 1
b = float(input("Введите переменную b: "))  # 2
condition_function = condition.function()
def f(x): return condition_function(a, b, x)


left_limit = float(input(
    "Введите переменную left_limit(левый предел рассматриваемой зоны): "))  # 1 # -0.5
right_limit = float(input(
    "Введите переменную right_limit(правый предел рассматриваемой зоны): "))  # 1.5 # 0.5

print(f"Левый предел: {left_limit}\nПравый предел: {right_limit}")


data_x = np.linspace(left_limit, right_limit, 1000)
data_y = [f(xi) for xi in data_x]
plt.plot(data_x, data_y)
data_y0 = [0 for _ in data_x]
plt.plot(data_x, data_y0)
plt.grid(True)
plt.show()
```

```python
accuracy = float(input("Введите точность: "))
print(f"Точность: {accuracy}")

iteration = int(input("Введите кол-во итераций: "))
print(f"Колличество итераций: {iteration}")
```

```python
print("Метод дихотомии")

x0 = 0.0
while True:
    x0 = float(
        input(f"Введите начальное приближение от {left_limit} до {right_limit}: "))

    if left_limit <= x0 <= right_limit:
        break

    print("Значение не находиться в диапозоне")

dichotomy_method(f, x0, (left_limit, right_limit), accuracy, iteration)
plt.title("Метод дихотомии")
plt.plot(data_x, data_y)
plt.plot(data_x, data_y0)
plt.grid(True)
plt.show()
```

```python
print("Метод хорд")

x0 = 0.0
while True:
    x0 = float(
        input(f"Введите начальное приближение от {left_limit} до {right_limit}: "))

    if left_limit <= x0 <= right_limit:
        break

    print("Значение не находиться в диапозоне")

chord_method(f, x0, accuracy, iteration)
plt.title("Метод хорд")
plt.plot(data_x, data_y)
plt.plot(data_x, data_y0)
plt.grid(True)
plt.show()
```

```python
print("Метод простых итераций")

g_function1 = condition.g_function1()
def g1(x): return g_function1(a, b, x)

g_function2 = condition.g_function2()
def g2(x): return g_function2(a, b, x)

f_function1 = condition.f_function1()
f_function2 = condition.f_function2()
def f1(x): return f_function1(a, x)
def f2(x): return f_function2(b, x)

data_y1 = [f1(xi) for xi in data_x]
data_y2 = [f2(xi) for xi in data_x]

x0 = 0.0
while True:
    x0 = float(
        input(f"Введите начальное приближение от {left_limit} до {right_limit}: "))

    if left_limit <= x0 <= right_limit:
        break

    print("Значение не находиться в диапозоне")

simple_iterations_method(
    f, g2, x0, accuracy, iteration)
plt.title("Метод простых итераций")
plt.plot(data_x, data_y)
plt.plot(data_x, data_y0)
plt.plot(data_x, data_y1)
plt.plot(data_x, data_y2)
plt.grid(True)
plt.show()
```

```python
print("Метод Ньютона")

def_function = condition.differential_function()
def def_f(x): return def_function(a, b, x)

x0 = 0.0
while True:
    x0 = float(
        input(f"Введите начальное приближение от {left_limit} до {right_limit}: "))

    if left_limit <= x0 <= right_limit:
        break

    print("Значение не находиться в диапозоне")


newton_method(f, def_f, x0, accuracy, iteration)
plt.title("Метод Ньютона")
plt.plot(data_x, data_y)
plt.plot(data_x, data_y0)
plt.grid(True)
plt.show()
```
