```python
from condition import *
from methods.euler import euler_method
from methods.rk4 import rk4_method
from solution import solution
from matplotlib import pyplot as plt
from prettytable import PrettyTable
```

```python
a = 0.5
b = 0.4
l = 1.9
n = 1

y0 = 1
interval = (0.0, 2.0)
h = 0.2
accuracy = 6
```

```python
print(function_as_string(a, b, l, n))

f = function(a, b, l, n)
fn = cauchy_function(a, b, l, n)

(data_x, data_y) = solution(interval, fn, h, accuracy)
```

```python
(data_x_euler, data_y_euler) = euler_method(interval, y0, f, h, accuracy)

plt.plot(data_x_euler, data_y_euler, label="Euler")
plt.plot(data_x, data_y, label="Original")
plt.legend()
plt.show()
```

```python
(data_x_rk4, data_y_rk4) = rk4_method(interval, y0, f, h, accuracy)

plt.plot(data_x_rk4, data_y_rk4, label="Runge-Kutta 4")
plt.plot(data_x, data_y, label="Original")
plt.legend()
plt.show()
```

```python
n = int(((interval[1] - interval[0]) / h)) + 1
bar = ["i"]
for i in range(n):
    bar.append(i)

t_data_x = ["xi"]
for i in range(n):
    t_data_x.append(round(data_x[i], 1))

t_data_y = ["yi"]
for i in range(n):
    t_data_y.append(data_y[i])

t_data_y_euler = ["yi Euler"]
for i in range(n):
    t_data_y_euler.append(data_y_euler[i])

t_data_y_rk4 = ["yi Runge-Kutta 4"]
for i in range(n):
    t_data_y_rk4.append(data_y_rk4[i])

table = PrettyTable(bar)
table.add_row(t_data_x)
table.add_row(t_data_y)
table.add_row(t_data_y_euler)
table.add_row(t_data_y_rk4)

print(table)
```
