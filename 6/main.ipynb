```python
import condition
import matplotlib.pyplot as plt
import numpy as np
from methods.dichotomy import dichotomy_method
from methods.golden_ratio import golden_ratio_method
from methods.fibonacci import fibonacci_method, fib
```

```python
left_limit = float(input("Введите левый предел: "))
right_limit = float(input("Введите правый предел: "))
accuracy = float(input("Введите точность: "))

print(f"Левый предел: {left_limit}")
print(f"Правый предел: {right_limit}")
print(f"Точность: {accuracy}")
```

```python
f = condition.function()

data_x = np.linspace(left_limit, right_limit, 1000)
data_y = [f(xi) for xi in data_x]

plt.plot(data_x, data_y)
plt.grid(True)
plt.show()
```

```python
xl1, xr1, iteration1 = dichotomy_method(f, (left_limit, right_limit), accuracy)
x1 = (xl1 + xr1) / 2

print(f"Кол-во итераций: {iteration1}")
print(f"Точность: {accuracy}")
print(f"Значение точки экстремума: {x1}")
print(f"Значение функции в этой точке: {f(x1)}")

plt.plot(data_x, data_y)
plt.grid(True)
plt.scatter(x1, f(x1), color="red")
```

```python
xl2, xr2, iteration2 = golden_ratio_method(f, (left_limit, right_limit), accuracy)
x2 = (xl2 + xr2) / 2

print(f"Кол-во итераций: {iteration2}")
print(f"Точность: {accuracy}")
print(f"Значение точки экстремума: {x2}")
print(f"Значение функции в этой точке: {f(x2)}")

plt.plot(data_x, data_y)
plt.grid(True)
plt.scatter(x2, f(x2), color="red")
```

```python
iteration = int(input("Введите колличество итераций для Фибаноччи: "))
xl3, xr3, iteration3 = fibonacci_method(f, (left_limit, right_limit), accuracy, iteration)
x3 = (xl3 + xr3) / 2

print(f"Кол-во итераций: {iteration3}")
if iteration3 == iteration:
    print("Достигнут предел итераций, возможно точность изменится")
    print(f"Точнось: {abs(int((xr3 - xl3) * 10**(abs(len(str(accuracy))) - 2)) / 10**(abs(len(str(accuracy))) - 2))}")
else:
    print(f"Точность: {accuracy}")

print(f"Значение точки экстремума: {x3}")
print(f"Значение функции в этой точке: {f(x3)}")

plt.plot(data_x, data_y)
plt.grid(True)
plt.scatter(x3, f(x3), color="red")
```
